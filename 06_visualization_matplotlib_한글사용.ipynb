```python
import numpy as np
import pandas as pd
from IPython.display import Image
import warnings
warnings.filterwarnings(action = "ignore")
```

```python
# 시각화 작업에 필요한 패키지와 라이브러리를 사용
import matplotlib as mpl
import matplotlib.font_manager as fm
import matplotlib.pyplot as plt
%matplotlib inline
# 한글폰트 사용시 마이너스 부호가 깨져 보이는 문제
mpl.rcParams["axes.unicode_minus"] = False
```

```python
# 그래프를 그리기 위해 더미 데이터를 만든다
data = np.random.randint(-100, 100, 50).cumsum()
print(data)
```

```python
plt.plot(range(50), data, "r")
plt.rcParams["figure.figsize"] = (10, 5) # 차트 크기 변경
plt.title("주식차트")
plt.xlabel("시간")
plt.ylabel("현재가")
plt.show() # 영어로 뜨는 정보를 숨김
```

```python
# 폰트를 설정해주기 앞서 설치된 matplotlib의 정보를 확인
print("버전 : ", mpl.__version__)
print("설치 위치 : ", mpl.__file__)
print("설정 위치 : ", mpl.get_configdir())
print("캐시 위치 : ", mpl.get_cachedir())
```

```python
# matplotlib의 위치 정보를 알았으니 해당 위치를 확인
print("설정 파일 위치 : ", mpl.matplotlib_fname())
```

```python
# 컴퓨터에 설치된 폰트 목록을 얻어오기
font_list = fm.findSystemFonts(fontpaths = None, fontext = "ttf")
print(len(font_list))
print(font_list[:10])
```

```python
# 나눔고딕코딩 글꼴을 사용할 예정이므로 이름에 "Nanum"이 들어간 폰트만 가져온다
# "Nanum"이 들어간 폰트이름 및 위치 프린트
[(f.name, f.fname) for f in fm.fontManager.ttflist if "Nanum" in f.name]
```

폰트를 사용하는 방법은 3가지

```python
# 1. FontProperties를 사용하는 방법 → 그래프의 폰트가 필요한 항목마다 지정해줘야 한다.
# 텍스트를 지정하는 항목에 한해서 사용할 수 있고 지정해준 항목에만 해당 폰트가 적용된다.
# matplotlib.pyplot → title(), xlabel(), ylabel(), legend(), text()
# matplotlib.axes → set_title()

# fname 옵션을 사용
path = "C:\\Windows\\Fonts\\나눔고딕코딩.ttf"
fontprop = fm.FontProperties(fname = path, size = 15)

plt.plot(range(50), data, "r")
plt.rcParams["figure.figsize"] = (10, 5) # 차트 크기 변경
plt.title("주식차트", fontproperties = fontprop)
plt.xlabel("시간", fontproperties = fontprop)
plt.ylabel("현재가", fontproperties = fontprop)
plt.show()
```

```python
# 2. matplotlib의 rcParams[]로 전역 글꼴 설정
# 기본설정 읽기
print("기본설정된 폰트크기 : ", plt.rcParams["font.size"])
print("기본설정된 폰트이름 : ", plt.rcParams["font.family"])
```

```python
# 기본으로 설정된 글꼴을 변경
plt.rcParams["font.size"] = 15
plt.rcParams["font.family"] = "NanumGothicCoding"
print("변경된 폰트크기 : ", plt.rcParams["font.size"])
print("변경된 폰트이름 : ", plt.rcParams["font.family"])
```

```python
plt.plot(range(50), data, "r")
plt.rcParams["figure.figsize"] = (15, 4) # 차트 크기 변경
plt.title("주식차트")
plt.xlabel("시간")
plt.ylabel("현재가")
plt.show()
```

```python
# 번외. rcParams 대신 FontProperties와 plt.rc를 사용하는 방법
path = "C:\\Windows\\Fonts\\나눔고딕코딩.ttf"
font_name = fm.FontProperties(fname = path, size = 18).get_name()
# print(font_name)
plt.rc("font", family = font_name)
```

```python
plt.plot(range(50), data, "r")
plt.rcParams["figure.figsize"] = (15, 4) # 차트 크기 변경
plt.title("주식차트")
plt.xlabel("시간")
plt.ylabel("현재가")
plt.show()
```
