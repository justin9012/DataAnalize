```python
# !# !pip install wordcloud
from wordcloud import WordCloud
```

```python
# read() 메소드로 워드클라우드로 표시할 데이터가 저장된 텍스트 파일을 읽어온다.
text = open("./data/wordData.txt").read()
text
```

```python
import random
```

```python
# word => 워드클라우드로 표시할 단어
# font_size => 워드클라우드로 표시할 단어의 크기
# position => 워드클라우드로 표시할 단어의 위치
# orientation => 워드클라우드로 표시할 단어의 배치
# random_state = None
# **hwargs => 워드 클라우드로 표시할 단어의 글꼴이 저장된 경로
# 워드클라우드를 그레이톤으로 표시하는 함수
def gray_color_func(word, font_size, position, orientation, random_state = None, **hwargs):
    # print(word, font_size, position, orientation)
    return "hsl(0, 0%%, %d%%)" % random.randint(60, 100)
```

```python
plt.figure(figsize = (15, 10))
plt.imshow(wordcloud.recolor(color_func = gray_color_func, random_state = 3), 
           interpolation = "bilinear")
plt.axis("off")
plt.show()
```

```python
import nltk
from konlpy.corpus import kolaw
from konlpy.tag import Twitter
t = Twitter()
```

```python
# konlpy 패키지는 대한민국 헌법 말뭉치인 kolaw와 국회법안 말뭉치인 kobill을 제공한다.
# constitution.txt는 konlpy 패키지에 내장된 파일이므로 파일 이름만 지정하면 읽어들일 수 있다.
ko_con_text = kolaw.open("constitution.txt").read()
ko_con_text
```

```python
# 한글 데이터 형태소 분석을 실행한다.
token_ko = t.nouns(ko_con_text)
token_ko
```

```python
# 불용어를 정의한다.
stop_words = ["제", "월", "일", "조", "수", "때", "그", "이", "바", "및", "안", "정", "위", "장",
             "관", "이상", "거나", "다만", "인", "모든", "경우", "기타", "항", "것", "둘"]
token_ko = [each_word for each_word in token_ko if each_word not in stop_words]
token_ko
```

```python
# 단어별 출현 빈도수를 계산한다.
ko = nltk.Text(token_ko)
ko.vocab().most_common(50)
```

```python
data = ko.vocab().most_common(500)
tmp_data = dict(data)
tmp_data
```

```python
# 딕셔너리에 저장된 데이터는 generate_from_frequencies() 메소드를 사용해서 워드클라우드를 실행해야
# 한다.
wordcloud = WordCloud(
    font_path = "c:/windows/fonts/나눔고딕코딩.ttf",
    background_color = "white",
    relative_scaling = 0.5,
).generate_from_frequencies(tmp_data)

plt.figure(figsize = (15, 10))
plt.imshow(wordcloud, interpolation = "bilinear")
plt.axis("off")
plt.show()
```
