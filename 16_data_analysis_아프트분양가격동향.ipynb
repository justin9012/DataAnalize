```python
# read_csv() / 파일이름이 한글로 되어 있으면 유니코드가 깨지고 못 읽어오는 현상 발생
# 제목을 영문으로 변경 후 읽어오거나 아래와 같이 encoding = "euc-kr"옵션을 지정
# 그래도 안된다면 위 옵션 뒤에 , engine = "python"까지 추가
per_sale = pd.read_csv("./data/전국_평균_분양가격_2018.6월_.csv", encoding = "euc-kr", engine = "python")
#per_sale = pd.read_csv("/data.perSale_2018_6.csv")
per_sale
```

```python
# 연도와 월은 연산에 사용되는 데이터가 아니라 구분에 사용되는 카테고리 형태의 데이터이므로 문자열 형태로 변환
per_sale["연도"] = per_sale["연도"].astype(str)
per_sale["월"] = per_sale["월"].astype(str)
```

```python
per_sale.dtypes
```

```python
# 분양가격의 데이터 타입을 숫자로 변경하고 평당 분양가격을 계산
# to_numeric() 메소드의 errors 속성을 "coerec"로 지정하면 오류가 발생된 데이트를 누락값으로 변경
per_sale["분양가격(㎡)"] = pd.to_numeric(per_sale["분양가격(㎡)"], errors = "coerce")
per_sale["평당분양가격(㎡)"] = per_sale["분양가격(㎡)"] * 3.3
```

```python
# usnull() 메소드와 sun() 메소드를 사용해서 누락값이 존재하는가 확인
print(per_sale.isnull().sum())
# missingno 라이브러리로 그래프를 그려서 누락값 확인하기 → 중간중간에 흰색으로 보이는 부분이 누락값을 의미
msno.matrix(per_sale, figsize = (10, 6))
plt.show
# per_sale = per_sale.dropna()
# print(per_sale.isnull().sum())
```

전국 규모별 평균 분양가격

```python
# 형식 → pivot_table(values, index, columns, aggfunc, fill_value) → values, index, colunms는 필수 입력 값 / 길게 나열되는 데이터를 index로 주는게 보기 좋다
# values : 통계 함수를 적용할 데이터프레임의 열 → 피벗 테이블 데이터
# index : 그룹화의 첫 번째 기준으로 사용되는 피벗 테이블로 가져올 데이터프레임의 열
# columns : 그룹화의 두 번째 기준으로 사용되는 피벗 테이블로 가져오 ㄹ데이터프레임의 열
# aggfunc : index와 colunms로 그룹화된 values에 적용할 함수를 문자열로 입력 / 생략가능(생략시 기본값은 평균을 계산하는 mean 함수)
# fill_value : NaN을 대체할 데이터를 입력 → 생략시 0이 기본값
per_sale.pivot_table("평당분양가격(㎡)", "규모구분", "연도", "min")
# per_sale.pivot_table("평당분양가격(㎡)", "규모구분", "연도", "mean")
# per_sale.pivot_table("평당분양가격(㎡)", "규모구분", "연도", "max")
```

```python
per_sale.pivot_table("평당분양가격(㎡)", ["연도", "월"], "규모구분")
```

```python
# 규모구분이 전체로 되어있는 금액으로 연도별 변동 금액을 살펴보자
# 규모구분이 전체인 데이터만 추출한다.
region_year_all = per_sale[per_sale["규모구분"] == "전체"]
region_year = region_year_all.pivot_table("평당분양가격(㎡)", "지역명", "연도").reset_index()
region_year
```

```python
region_year["변동액"] = region_year["2018"] - region_year["2015"]
max_delta_price = np.max(region_year["변동액"]) * 1000
min_delta_price = np.min(region_year["변동액"]) * 1000
mean_delta_price = np.mean(region_year["변동액"]) * 1000

print("2015년부터 2018년까지 분양가는 계속 상승했으며, 상승액이 가장 큰 지역의 ", end = "")
print("제주로 상승액은 평당 {:,.0f}원 이다.".format(max_delta_price))
print("상승액이 가장 적은 지역은 울산이며 상승액은 평당 {:,.0f}원 이다.".format(min_delta_price))
print("전국 평균 변동액은 평당 {:,.0f}원 이다.".format(mean_delta_price))
region_year
```
