```python
import folium
```

```python
# Map() 메소드로 표시할 지도의 범위를 지정한다.
# location 속성으로 위도와 경도를 지정해서 표시되는 지도의 정중앙 좌표를 설정하고 zoom_start 속성
# 으로 지도의 배율을 설정한다.
# tiles 속성으로 지도 스타일을 지정한다. Open Street Map이 기본값으로 사용되고 Stamen Terrain,
# Stamen Toner를 사용할 수 있다.
# folium.Map(location = [위도, 경도], zoom_start = 배율)
map_folium = folium.Map(location = [37.5609615, 126.9745681], zoom_start = 15, 
                        tiles = "Open Street Map")
# folium의 save() 메소드로 작성된 지도 차트를 html 파일로 저장할 수 있다.
map_folium.save("./output/map1.html")
map_folium
```

```python
map_folium = folium.Map(location = [37.5609615, 126.9745681], zoom_start = 14)

# Folium으로 표현되는 지도에 표시할 Marker를 만든다. => Marker를 만든 후 add_to() 메소드로 지도에
# 추가한다.
# Marker() 메소드로 특정 위치를 아이콘으로 강조할 수 있다.
# location 속성에 마커를 표시할 위치를 지정하고 popup 속성에는 마커를 클릭했을 때 메시지로 표시할
# 문자열을 지정하고 icon 속성에 Icon() 메소드로 color 속성에 마커의 색상과 icon 속성에 아이콘
# 종류를 지정한다.
# 마커 위에 표시되는 아이콘은 https://getbootstrap.com/docs/3.3/components/#glyphicons-glyph에서
# 제공하는 이미지를 표시할 수 있다.
folium.Marker(location = [37.5609615, 126.9745681], popup = "서울특별시청",
             icon = folium.Icon(color = "red", icon = "star")).add_to(map_folium)
folium.Marker(location = [37.5657852,126.9740246], popup = "덕수궁",
             icon = folium.Icon(color = "green", icon = "ok")).add_to(map_folium)
folium.Marker(location = [37.5748406,126.9697778], popup = "서울지방경찰청",
             icon = folium.Icon(color = "purple", icon = "chevron-down")).add_to(map_folium)
folium.Marker(location = [37.5759369,126.974627], popup = "광화문",
             icon = folium.Icon(color = "pink", icon = "pushpin")).add_to(map_folium)
folium.Marker(location = [37.5631549,126.9695791], popup = "영풍문고",
             icon = folium.Icon(color = "orange", icon = "download-alt")).add_to(map_folium)

map_folium.save("./output/map4.html")
map_folium
```

```python
map_folium = folium.Map(location = [37.5609615, 126.9745681], zoom_start = 14)
folium.Marker(location = [37.5609615, 126.9745681], popup = "서울특별시청",
             icon = folium.Icon(color = "red", icon = "star")).add_to(map_folium)
folium.CircleMarker(location = [37.5657852,126.9740246], popup = "덕수궁",
             radius = 100, color = "#FF00FF", fill_color = "#00FFFF", 
                   fill = True).add_to(map_folium)
map_folium.save("./output/map5.html")
map_folium
```
